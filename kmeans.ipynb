```python
#k-means
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)
kmeans=KMeans(n_clusters=4)
kmeans.fit(X)
```

```python
y_pred=kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap= 'viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=50, c='pink', marker='o')
plt.title('K-means Clustering') 
plt.xlabel('Feature 1')
plt.ylabel('Feature 2') 
plt.show()
```

```python
#2 exercise, find nbr cluster
wss=[]
krange = range (1,11)
for k in krange :
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(X)
    wss.append(kmeans.inertia_)
    
```

```python
plt.figure(figsize=(10,5))
plt.plot(krange,wss, marker='o', linestyle='--')
plt.show()
```

```python
#normaisation des pixel 
X= X / 255.0
```

```python
#FIND NBR CLUSTER 
wss=[]
krange = range (1,15)
for k in krange :
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(X)
    wss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(krange,wss, marker='o', linestyle='--')
plt.xlabel('Nombre de clusters (k)')
plt.title('Méthode du Coude pour trouver le k ')
plt.show()
```
