```python
import pandas as pd
# 리스트를 사용해서 시리즈
data = [10, 20, 30, 40]
series1 = pd.Series(data)
print(series1)

# 인덱스 지정해 Series생성
indexT = ['a', 'b','c','d']
#pd.Series(데이터, 인덱스)
Series2 = pd.Series(data, index=indexT)
print(Series2)

#딕셔너리로 데이터 생성
dic = {'a':1, 'b':2, 'c':3}
Series3 = pd.Series(dic) #data.dic : 즉 점(.)은 보통 “이 객체 안의 무엇”이라는 뜻이야.
print(Series3)
```

```python
import pandas as pd
# 리스트를 사용해서 시리즈
da_list = ['kim', 25,'m'], ['lee', 20, 'f'], ['jeu', 22, 'f']
daframe = pd.DataFrame(da_list)#그런데 열 이름을 따로 안 줬으니까 판다스가 자동으로
print(daframe)

Add_colum = ['name', 'age', 'sex'] #그 열 이름으로 쓰고 싶은 목록이야.
daframe2 = pd.DataFrame(da_list, Add_colum)
#pd.DataFrame(data, index=..., columns=...)즉 두 번째 위치에 그냥 Add_colum을 넣으면,
#판다스가 그걸 행 인덱스(index) 쪽으로 해석할 수 있어
print(daframe2)

daframe3= pd.DataFrame(da_list, columns = Add_colum) #이터프레임의 열 이름을 Add_colum으로 지정해라
# columns=Add_colum 이렇게 써야 판다스가:“아, 이건 열 이름 목록이구나 라고 알아들어.” columns=Add_colum

print(daframe3)

#딕셔너리로 데이터 생성
da_di = { 'name' : ['kim', 'lee', 'jeu'],
         'age'  : [ 25, 20, 22] ,
          'sex' : ['m', 'f', 'f']
          }
dicto = pd.DataFrame(da_di)
print(dicto)

```

```python
#앞서 딕셔너리도 데이터 생성
da_di = { 'name' : ['kim', 'lee', 'jeu'],
         'age'  : [ 25, 20, 22] ,
          'sex' : ['m', 'f', 'f']
          }
dicto = pd.DataFrame(da_di)
print(dicto)
#열선택  - 이건 결과: 시리즈(Series)처럼 나
print(dicto['name'])

#2개 이상 열 선택 열 여러 개를 고를 때는 리스트로 묶어야 해.
#여기서 바깥 대괄호 []는 데이터프레임 선택이고,안쪽 대괄호 ['name', 'age']는 열 이름들의 리스트야.
#결과: 데이터프레임(DataFrame)
#즉 “데이터프레임에서 이 열 이름 리스트를 선택해라”

print(dicto[ ['name', 'age']  ] )

#행 row선택
print(dicto[ :2]) # 0번째줄인 첫번째 줄에서 1번째행인 두번째 행선택
```

```python
#라벨 기반선택인 loc함수
da_di = { 'name' : ['kim', 'lee', 'jeu'],
         'age'  : [ 25, 20, 22] ,
          'sex' : ['m', 'f', 'f']
          }
addindex = ['a', 'b','c']
dicto = pd.DataFrame(da_di, index= addindex)
print(dicto)

#df.loc[행, 열]*********
#df.loc[행] : 특정 행 선택
# df.loc[:, 열] : 특정 열 전체 선택
#df.loc[행, 열] : 특정 행과 열의 교차점 선택

#라벨 기반선택인 loc함수로 a행만 선택
print(dicto.loc['a'])

#라벨 기반선택인 loc함수로 특정 열선택
print(dicto.loc['age' : 'sex']) #쉼표가 없으면 loc[...]의 한 자리 입력은 행 기준으로 처리된다는 점

#열을 고르려면 행 자리와 열 자리를 나눠서 써야 해.
#: → 모든 행 - 'age':'sex' → age부터 sex까지 열 선택
print(dicto.loc[ :, 'age' : 'sex'])

#라벨 기반선택인 loc함수로 특정 열선택
#print(dicto.loc['sex']) #sexs는 열이름 맨앞에는 행을 꼭써줘야됨
#행 인덱스: a, b, c  열 이름: name, age, sex
#: → 모든 행 이고 'sex' → sex 열

print(dicto.loc[ : , 'sex'])


#특정행과 열선택
print(dicto.loc ['b','age'])

#특정행과 열선택
print(dicto.loc ['a','age'],['c','name'])
#"dicto에서 'a'행의 'age' 값을 가져와!" → 25 (성공)
#['c', 'name']: "그리고 그냥 이 글자 리스트를 출력해!" → ['c', 'name'] (데이터프레임과 상관없는 그냥 글자 묶음)

#여러 개를 고를 때는 대괄호 [ ]로 보따리를 만든다!"
#df.loc[[행1, 행2], [열1, 열2]]

#print(dicto.loc [['a','age'], ['c','name'] ] ) 에러 행자리에 age는 없은 열자리에도 name은 없음
#loc 안의 쉼표(,)를 기준으로 왼쪽은 무조건 행(Row), 오른쪽은 무조건 열(Column) 보따리를 넣어야 합니다.
print(dicto.loc [ ['a','c'], ['name', 'age']])

#loc[ : 자, 이제 데이터를 찾을 거야.['a', 'c'] : (행 자리) 'a'랑 'c' 행을 선택해줘.
#, : (벽) 이제부터는 열을 말해줄게.
#['age', 'name'] : (열 자리) 'age'랑 'name' 열을 선택해줘.
#] : 끝!

```

```python
#특정행과 열선택
print(dicto.loc ['b','age'])

#특정행과 열선택
print(dicto.loc ['a','age'],['c','name'])
```

```python
#iloc 함수는 행과 열의 위치를 기준으로 대이터 선택
print(dicto)

#특정행 선택
print(dicto.iloc[1])

# 특정 열 선택
print(dicto.iloc[ : , 2])

# 특정행과 열선택
print(dicto.iloc[ 1, 1])

#여러개의 행과 열 선택
#print(dicto.iloc [ 1, 1], [0,2])
print(dicto.iloc [[ 1, 2], [0,0]])
print(dicto.iloc [[ 1, 2], [1,0]])
```

```python
print(dicto)

#조건기반 선택 필터링
print(dicto['age'] >20) #맞다 틀리다만 보는 것 즉 데이터를 가져오는게 아님

print(dicto[['age'] ]>20) #*'결과물의 모양'**을 표 형태로 유지하고 싶을 때 리스트 []를 사용해 한 번 더 감싸주는 것입니다.

#조가 가장 헷갈리실 텐데, 여기서 바깥쪽 대괄호 []는 **"진짜 데이터를 꺼내는 입구"**
#안쪽 dicto['age'] > 20: 아까 만든 **'필터지'**입니다. (True/False가 적힌 종이)
#바깥쪽 dicto[ ... ]: 이 필터지를 대입해서 **True인 행만 통과시켜!**라는 명령어입니다.

#dicto['age'] : 'age'라는 열 하나를 선택할 때 (집게)
#dicto[['age']] : 'age'를 담은 리스트를 선택할 때 (보따리 집게)
#dicto[조건식] : 조건이 참(True)인 행만 필터링할 때 (필터 집게)

print(dicto[dicto['age'] >20])

#여러조건 검색
#에러 print(dicto [dicto['age'] > 20] & [dicto['sex'] == 'f'])
#에러 print(dicto [dicto['age'] > 20]) & ([dicto['sex'] == 'f'])

# 전체 데이터를 담은 dicto라는 바구니 [ 그 안에 조건을 넣으세요 ]
# 조건 1: (dicto['age'] > 20)
# 조건 2: (dicto['sex'] == 'f')
# print(dicto [dicto['age'] > 20]) & ([dicto['sex'] == 'f'])

#에러 print(dicto ([dicto['age'] > 20])) & ([dicto['sex'] == 'f'])

#**"Pandas에서 대괄호 []는 필터링할 때 '입구'와 '출구' 역할을 하는 거대한 울타리"**라고 생각하면 됩니다.
#dicto [  ]  <-- 이게 기본 틀이에요.

#2단계 그 울타리 안에 우리가 원하는 조건을 넣을 거예요.
 #"나이가 20보다 커?" → dicto['age'] > 20
 # "성별이 여자야?" → dicto['sex'] == 'f'

#3단계 조건이 두 개 이상일 때는 컴퓨터가 헷갈리지 않게 소괄호 ( )로 각각 감싼 뒤에 & (그리고)나 | (또는)로 붙여줍니다.
#(dicto['age'] > 20) & (dicto['sex'] == 'f')
print ( dicto [ (dicto['age'] > 20 ) & ( dicto['sex'] == 'f' ) ] )


#특정 열의 값이 특정리스트에 포함이 안될때
print ( dicto [ dicto ['name'].isin( [ 'kim', 'lee' ] ) ] )
```

```python
#sort_values() 와 sort_index()함수 사용해 데이터 정렬

import pandas as pd
#에러 df = pd.DataFrame{ ['name' : 'kim','lee','jun' ],
                  #['age' : 22, 25 , 20],
                   #['score': 100, 50, 99]
                   #}


#딕셔너리(Dictionary)**라는 형식을 써야 하는데, 이게 바로 { 키 : 값 }
#중괄호 {} 안에 '열 이름' : [데이터 리스트] 순서로 넣어야 합니다.
""" 에러 df = pd.DataFrame { 'name' : ['kim', 'lee', 'jun'],
                   'age' : [22, 25, 20],
                    'score' : [ 100, 50, 99]}
"""
#함수 실행은 소괄호 (): pd.DataFrame은 함수예요. 함수를 실행해서 결과를 만들 때는 반드시 **소괄호 ( )**로 감싸야 합니다.

#(소괄호): pd.DataFrame(...) 처럼 함수를 실행할 때 재료를 던져주는 입구입니다.

#{ } (중괄호): "이름표(Key)와 내용물(Value)을 연결할게!" 라는 뜻입니다.
#예: 'name' : ['kim', 'lee', 'jun'] (이름표는 name이고, 내용은 이 리스트야!)

#[ ] (대괄호): "여러 개의 데이터를 한 묶음으로 담을게!" 라는 뜻입니다.
#예: [22, 25, 20] (나이 세 개를 한 보따리에 담았어!)
#예: 'name' : ['kim', 'lee', 'jun'] (이름표는 name이고, 내용은 이 리스트야!)
df = pd.DataFrame ( { 'name' : ['kim', 'lee', 'jun'],
                   'age' : [22, 25, 20],
                    'score' : [ 100, 50, 99] } )
#**'name', 'age', 'score'**는 **열(Column)**의 이름이 되고,
#['kim', 'lee', 'jun'] 같은 리스트들은 그 열의 데이터가 됩니다.

print(df)
# 점수를 기준으로 내림차순 높은 순으로 정리
df_sorted = df.sort_values(by='score', ascending=False)
print(df_sorted)

# 점수를 기준으로 오름차순정리
df_sortedH = df.sort_values(by = 'score')
print(df_sortedH)

```

```python
#필터링
#에러 filter_df = df [ df['age'> = 21]]

#df['age']: "df야, 'age'라고 적힌 세로 기둥(열) 좀 가져와봐."
#df['age'] >= 21: "그 기둥에 적힌 숫자들 중에 21 이상인 것들만 골라서 True/False로 표시해줘."
#df[ ... ]: "방금 표시한 **True(참)인 줄(행)**만 가로로 쏙쏙 뽑아서 새 표를 만들어줘!"


filter_df = df [ df['age'] >= 21 ]
print(filter_df)

#여러 조건
flter_dfS = df [   (df['age'] >=21) &  ( df ['score'] > 50 )]
print(flter_dfS)

#특정행과 열을 선택
resF = df [ ['name', 'score']]
print(resF)
```
